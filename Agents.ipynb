```python
from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from tools import *
```

```python
load_dotenv()
OPENAI_API_KEY = os.getenv("open_api_key")
OPENAI_BASE_URL = os.getenv("openApiBaseUrl")  # optional
```

```python
search_web.invoke("what is machine learning?")
```

```python
llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0,
    openai_api_key=os.getenv("open_api_key") ,
    base_url=os.getenv("openApiBaseUrl")
)
```

```python
agent = create_agent( llm , tools=[search_web,Calculate])
```

```python
res = agent.invoke({
    "messages": [{"role": "user", "content": "who won world cup 2022?"}]
})
print(res)
```
